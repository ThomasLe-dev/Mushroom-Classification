```python
import pandas as pd
#import thư viện xử lý data từ file csv
```

```python
data = pd.read_csv('mushrooms.csv')
#gán biến data cho dữ liệu đọc từ file csv
```

```python
data = data.astype('category')
#Convert raw data từ object sang category để có thể thao tác bằng encoder
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for column in data.columns:
    data[column]=le.fit_transform(data[column])
#Convert data sang dạng số(loại dữ liệu mà máy có thể hiểu được)
```

```python
data.head()
#check xem dữ liệu đã được converse hay chưa
```

```python
X = data.drop('type', axis = 1)
#X tập chứa data sau khi loại bỏ cột phân loại
```

```python
y = data['type']
#y tập chứa data phân loại của nấm
```

```python
X
```

```python
from sklearn.decomposition import PCA

pca1 = PCA(n_components = 7)
pca_fit1 = pca1.fit_transform(X)

#áp dụng kỹ thuật pca nhằm giảm kích thước của data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(pca_fit1,y,test_size=0.20,
                                               random_state=42)

#chia data ra thành 2 bộ dữ liệu, bộ train và bộ test gồm: 80% để huấn luyện, 20% dùng để kiểm tra
# sử dụng random_state để trộn tập dữ liệu 42 lần
```

```python
from sklearn.svm import SVC
#import thuật toán phân loại svm
```

```python
svc = SVC()
# khai báo mẫu sử dụng SVC
svc.fit(X_train,y_train)
# tiến hành huấn luyện mẫu vừa khai báo
# bằng tập dữ liệu huấn luyện
```

```python
#Xử lý dự đoán về độ chính xác của thuật toán
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
#import thư viện xử lý độ chính xác của thuật toán
```

```python
print("Accuracy SVM",accuracy_score(y_test,y_pred)*100)
#in ra độ chính xác của thuật toán dựa trên bộ dữ liệu test
```

```python
svm_model = SVC()
svm_model.fit(pca_fit1,y)
#Lưu lại model thuật toán
```
